```python
def ip_divide(x):
    ret = list()
    for i in range(len(x)):
        temp = list(map(int, x[i].split('.')))
        ret.append(temp)

    return ret
```

```python
prot_a_srcip = ip_divide(np.array(prot_a['src_ip']))
prot_a_dstip = ip_divide(np.array(prot_a['dst_ip']))

prot_b_srcip = ip_divide(np.array(prot_b['src_ip']))
prot_b_dstip = ip_divide(np.array(prot_b['dst_ip']))

prot_c_srcip = ip_divide(np.array(prot_c['src_ip']))
prot_c_dstip = ip_divide(np.array(prot_c['dst_ip']))
```

```python
def insert_newip(data, x, y):
    new_data1 = pd.DataFrame(x, columns = ['0', '1', '2', '3'])
    new_data2 = pd.DataFrame(y, columns = ['4', '5', '6', '7'])

    new_data1 = new_data1.loc[~new_data1.index.duplicated(keep='first')]
    new_data2 = new_data2.loc[~new_data2.index.duplicated(keep='first')]
    data = data.loc[~data.index.duplicated(keep='first')]

    data = pd.concat([data, new_data1, new_data2], axis = 1)

    return data
```

```python
prot_a = insert_newip(prot_a, prot_a_srcip, prot_a_dstip)
prot_b = insert_newip(prot_b, prot_b_srcip, prot_b_dstip)
prot_c = insert_newip(prot_c, prot_c_srcip, prot_c_dstip)
```

```python
from scipy.spatial.distance import cdist
```

```python
def make_B(centroids):
    distances = cdist(centroids, centroids, metric = 'euclidean')
    ssd = distances ** 2

    return list(ssd)
```

```python
def make_W(n_clusters, labels, centroids, data):
    ssd = list()
    for i in range(n_clusters):
        cluster_points = data[data['labels'] == i]
        cluster_points.drop('labels', axis = 1, inplace = True)
        cluster_points = np.array(cluster_points)
        ssd_temp = np.sum((cluster_points - centroids[i]) ** 2, axis = 1)
        ssd.append(ssd_temp)

    return ssd
```

```python
x = list()
y = list()

for i in range(2, 21):
    x.append(i)
    km = KMeans(n_clusters = i, init = 'k-means++')
    km.fit(scaled_data)
    centroids = km.cluster_centers_
    labels = km.labels_
    data_temp = pd.DataFrame(scaled_data)
    labels = pd.DataFrame(labels.reshape(-1, 1), columns = ['labels'])
    data_temp = pd.concat([data_temp, labels], axis = 1)
    B = make_B(centroids)
    W = make_W(i, labels, centroids, data_temp)
    W_st = 0
    B_st = 0
    for i in range(len(W)):
        W_st += sum(W[i])
    for i in range(len(B)):
        for j in range(len(B[0])):
            if i != j:
                B_st += B[i][j]

    y.append((B_st / (i - 1)) / (W_st / (20 - i)))
```
